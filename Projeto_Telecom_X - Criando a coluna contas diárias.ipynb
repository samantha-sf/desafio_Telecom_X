```python
import pandas as pd
import requests
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
```

#Extraindo os dados da API

```python
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)
df
```

#Normalizando colunas aninhadas

```python
df_plano = pd.json_normalize(data)
df_plano.head(10)
```

# Verificando os valores das colunas

```python
for col in df_plano.columns:
  print(f"valores únicos na coluna'{col}': {df_plano[col].nunique()}")
  if df_plano[col].nunique() < 50: # Imprima os valores apenas se não forem muitos.
    print(df_plano[col].unique())
    print('-' * 50)
```

#Verificando se há valores nulos

```python
print("Número de nulos", df_plano.isnull().sum())
```

#Verificando se há linhas duplicados

```python
print("Numero de duplicados", df_plano.duplicated().sum())
```

#Retirando espaços e verificando se há strings vazias

```python
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

#Alterando a coluna "account.Charges.Total" para Float

```python
df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
print(df_plano['account.Charges.Total'].dtype)
```

#Eliminando as linhas vazias de Churn



```python
df_plano = df_plano[df_plano['Churn'].str.strip() != '']
print("Número de linhas depois de tirar os vazios da coluna 'Churn':", len(df_plano))
```

```python
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```
