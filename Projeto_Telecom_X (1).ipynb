```python
import pandas as pd
import requests
```

```python
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
```

#Extraindo os dados da API

```python
response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)
df
```

#Normalizando colunas aninhadas

```python
df_plano = pd.json_normalize(data)
df_plano.head(10)
```

# Verificando os valores das colunas

```python
for col in df_plano.columns:
  print(f"valores únicos na coluna'{col}': {df_plano[col].nunique()}")
  if df_plano[col].nunique() < 50: # Imprima os valores apenas se não forem muitos.
    print(df_plano[col].unique())
    print('-' * 50)
```

#Verificando se há valores nulos

```python
print("Número de nulos", df_plano.isnull().sum())
```

#Verificando se há linhas duplicados

```python
print("Numero de duplicados", df_plano.duplicated().sum())
```

#Retirando espaços e verificando se há strings vazias

```python
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

#Alterando a coluna "account.Charges.Total" para Float

```python
df_plano['account.Charges.Total'] = pd.to_numeric(df_plano['account.Charges.Total'], errors='coerce')
print(df_plano['account.Charges.Total'].dtype)
```

#Eliminando as linhas vazias de Churn



```python
df_plano = df_plano[df_plano['Churn'].str.strip() != '']
print("Número de linhas depois de tirar os vazios da coluna 'Churn':", len(df_plano))
```

```python
df_plano.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

#Criando uma coluna: Contas diárias. Utilizando o faturamento mensal para calcular o valor diário.

```python
df_plano
```

```python
df_plano['Total.Day'] = ((df_plano['account.Charges.Total'] / df_plano['customer.tenure']) /30).round(2)
df_plano.insert(19, 'Total.Day', df_plano.pop('Total.Day'))
df_plano
```

#Padronização e Transformação de Dados

```python
print(df_plano.columns.tolist())
```

```python
df_plano = df_plano.rename(columns={
    'customerID': 'Id cliente',
    'Churn': 'Desistente',
    'customer.gender': 'Sexo',
    'customer.SeniorCitizen': 'Idade 65+',
    'customer.Partner': 'Casado',
    'customer.Dependents': 'Filhos',
    'customer.tenure': 'Há quanto tempo é cliente (meses)',
    'phone.PhoneService': 'Serviço de telefone',
    'phone.MultipleLines': 'Contratou mais de uma linha',
    'internet.InternetService': 'Serviço de internet',
    'internet.OnlineSecurity': 'Segurança online',
    'internet.OnlineBackup': 'Backup online',
    'internet.DeviceProtection': 'Proteção do dispositivo',
    'internet.TechSupport': 'Suporte técnico',
    'internet.StreamingTV': 'TV streaming',
    'internet.StreamingMovies': 'Streaming de Filmes',
    'account.Contract': 'Duração do contrato',
    'account.PaperlessBilling': 'Fatura digital',
    'account.PaymentMethod': 'Forma de pagamento',
    'Total.Day': 'Valor diário',
    'account.Charges.Monthly': 'Valor mensal',
    'account.Charges.Total': 'Faturamento total'
})
```

```python
print(df_plano.columns.tolist())
```

```python
df_plano.head(10)
```

```python
clientes_atuais = (df_plano['Desistente'] == 'No').sum()
print(f"O número de clientes atuais é {clientes_atuais}")
clientes_desistentes = (df_plano['Desistente'] == 'Yes').sum()
print(f"O número de clientes desistentes é {clientes_desistentes}")
percentual_desistentes = (clientes_desistentes * 100 / (clientes_desistentes + clientes_atuais))
print(f"O percentual de clientes desistentes é {percentual_desistentes:.2f}%")
```

#Análise descritiva: média, mediana e desvio padrão

#Média

```python
media_tempo_contrato = df_plano['Há quanto tempo é cliente (meses)'].mean()
print(f"A maioria dos clientes tem {media_tempo_contrato:.0f} meses de contrato.")
```

```python
media_gasto_mensal = df_plano['Valor mensal'].mean()
print(f"A média de gastos mensais dos clientes é de R${media_gasto_mensal:.2f}")
```

```python
media_gasto_diario = df_plano['Valor diário'].mean()
print(f"A média de gastos diários dos clientes é de R${media_gasto_diario:.2f}")
```

```python
media_gasto_mensal_desistente = df_plano[df_plano['Desistente'] == 'Yes']['Valor mensal'].mean()
print(f"A média de gastos mensais dos clientes desistentes é de R${media_gasto_mensal_desistente:.2f}")
```

```python
media_gasto_mensal_nao_desistente = df_plano[df_plano['Desistente'] == 'No']['Valor mensal'].mean()
print(f"A média de gastos mensais dos clientes não desistentes é de R${media_gasto_mensal_nao_desistente:.2f}")
```

#Mediana

```python
mediana_tempo_contrato = df_plano['Há quanto tempo é cliente (meses)'].median()
print(f"A mediana dos clientes tem {mediana_tempo_contrato:.0f} meses de contrato.")
```

#Desvio padrão

```python
df_plano['Há quanto tempo é cliente (meses)'].std()
```

#Distribuição da Evasão: gráfico para analisar a proporção de clientes que permaneceram e os que saíram.

```python
df_plano.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
df_plano['Desistente'] = df_plano['Desistente'].str.strip()
```

```python
df_plano['Desistente'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightgreen', 'lightcoral'], labels=['Não desistentes', 'Desistentes'], startangle=90)
plt.title('Proporção de desistência de clientes')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data=df_plano, x='Há quanto tempo é cliente (meses)', hue='Desistente', palette='Set2')
plt.title('Desistência de clientes por tempo de contrato')
plt.xlabel('Tempo de contrato (meses)')
plt.ylabel('Número de clientes')
plt.legend(title='Desistente', loc='upper right')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
